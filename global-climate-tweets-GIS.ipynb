```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from itables import show
import os
import glob
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
tqdm.pandas()
import json
files = sorted(glob.glob("data/hydrated/climate_tweets_*.jsonl"))
print(files)
pd.set_option('display.max_colwidth', -1)
```

```python
points = []
for filename in tqdm(files):
    with tqdm(total = os.path.getsize(filename)) as pbar:
        with open(filename, "r") as file:
            for line in file:
                pbar.update(len(line))
                row = json.loads(line)
                if row["coordinates"]:
                    points.append(row["coordinates"]["coordinates"])
print(len(points))
```

```python
display(points[1:10])
df = pd.DataFrame(points, columns=["lng", "lat"])
df.to_csv('climate_tweet_latlngs.csv', index=False)
```

```python
import folium
from folium.plugins import HeatMap

map = folium.Map(location = [0,0], zoom_start = 2)
HeatMap([[lat, lng] for lng, lat in points]).add_to(map)
map.save("climate_tweet_heatmap.html")
map
```
